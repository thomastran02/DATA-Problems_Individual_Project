## Data preprocessing

```python
f = open('challenge_set.json','r')
data = json.loads(f.read())
df = pd.json_normalize(data,record_path=['playlists'],meta=['date'])

tracks = df['tracks']

def new_columns(key,col = tracks,empty = ['Empty']):
    info = []
    for i in col:
        if i==[]:
            info.append(empty)
        else:
            lst = list()
            for j in i:
                lst.append(j[key])
            info.append(lst)
    return info

keys = ['track_name','artist_name','track_uri','artist_uri','album_uri','duration_ms','album_name']
for k in keys:
    df[k] = new_columns(k)
```

```python
def unique_tracks(playlist):
    ret_lst = list()
    for i in playlist:
        for j in i:
            if j not in ret_lst:
                ret_lst.append(j)
    return ret_lst
```

```python
def make_dict(column,values,full_track,data_frame):
    dictionary = dict()
    for track in full_track:
        for index in data_frame.index:
            if track in data_frame[column][index]:
                i = data_frame[column][index].index(track)
                dictionary[track]=data_frame[values][index][i]
                break
    return dictionary
```

### This creates an implicit rating of each track by assigning each track a rating by the amount of times the specific artist appears in the playlist

```python
def create_ratings(user,tracks,track_col='track_uri',art_col='artist_uri'):
    lst = list()
    for track in tracks:
        if track in user[track_col]:
            lst.append((user[art_col].count(track_artist_uri[track]))/len(user[art_col]))
        else:
            lst.append(0.0)
    return lst
```

### This will create the user-item rating matrix

```python
def create_matrix(data_frame,tracks,track_col='track_uri',art_col='artist_uri'):
    lst = list()
    for index in range(len(data_frame)):
        lst.append(create_ratings(data_frame.iloc[index],tracks,track_col,art_col))
    arr = np.array(lst)
    return arr
```

```python
# Get top k users
def get_top_users(userid,similar,k=10,threshold=0):
    matrix = similar.drop(userid)
    similar_users = matrix[matrix[userid]>threshold].sort_values(by=[userid],ascending=False)[userid].head(k)
    return similar_users

# Tracks target user listened to
def tracks_in_playlist(tracks,userid):
    track_user_has = tracks[tracks.index==userid]
    track_user_has = track_user_has.loc[:,(track_user_has!=0).any(axis=0)]
    return track_user_has

# Similar User Tracks
def track_in_similar(similar_users,tracks):
    similar_user_tracks = tracks[tracks.index.isin(similar_users.index)].replace(0,np.nan).dropna(axis=1,how='all')
    similar_user_tracks = similar_user_tracks.fillna(0)
    return similar_user_tracks
```

```python
def recommend(userid,tracks,similar,k=10,threshold=0):
    # Get top k users
    similar_users = get_top_users(userid,similar,k,threshold)
    
    # Tracks target user listened to
    track_user_has = tracks_in_playlist(tracks,userid)
    
    # Similar User Tracks
    similar_user_tracks = track_in_similar(similar_users,tracks)
    
    # Drop tracks from similar_user_track that were already watched by user
    similar_user_tracks.drop(track_user_has.columns,axis=1,inplace=True,errors='ignore')
    
    # A dictionary to store item scores
    item_score = {}
    
    # Recommend Items
    for i in similar_user_tracks.columns:
        # Get rating for track i
        track_rating = similar_user_tracks[i]
        
        # Create variable to store score
        total = 0
        
        # Create variable to store number of scores
        count = 0
        
        # Loop through similar users
        for u in similar_users.index:
            # If track has rating
            if track_rating[u]:
                # Score is sum of user similarity score multiply by track rating
                score = similar_users[u]*track_rating[u]
                
                # Add score to total score for the movie so far
                total+=score
                # Add 1 to count
                count+=1
        
        # Average score for item
        item_score[i]=total/count
    
    item_score = pd.DataFrame(item_score.items(),columns=['track','track_score'])
    
    # Sort items by score
    ranked_item_score = item_score.sort_values(by='track_score',ascending=False)
    
    return ranked_item_score
```

```python
temp = recommend(900,df_matrix_raw,user_similarity,k=5)
```

```python
def precision(temp,true):
    count = 0
    num_rec_tracks = len(temp)*len(temp[0])
    for user in range(len(temp)):
        for track in temp[user]:
            if track in true[user]:
                count+=1
    return count/num_rec_tracks
```
