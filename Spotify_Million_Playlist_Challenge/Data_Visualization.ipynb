```python
def Convert(string):
    string = string.replace('[',"")
    string = string.replace(']',"")
    string = string.replace("'","")
    l = list(string.split(', '))
    return l

def Column(column):
    lst = list()
    for c in column:
        lst.append(Convert(c))
    return lst
```

### This step will take a subset of the artists and count how many times each of these artists appear in a playlist together

```python
from IPython.display import HTML, display
import tabulate
```

This gives a probability given one artist is in the playlist

```python
def count_same_playlist(lst,playlist):
    ret_lst = list()
    for i in lst:
        elmt = [i]
        total=0
        for j in playlist:
            if i in j:
                total+=1
        for k in lst:
            counter=0
            for l in playlist:
                if i in l and k in l:
                    counter+=1
            elmt.append(counter*100/total)
        ret_lst.append(elmt)
    return ret_lst
```

This gives probability of both artist for all the playlist

```python
def count_same_playlist1(lst,playlist):
    ret_lst = list()
    for i in lst:
        elmt = [i]
        total=0
        for j in playlist:
            if i in j:
                total+=1
        for k in lst:
            counter=0
            for l in playlist:
                if i in l and k in l:
                    counter+=1
            elmt.append(counter*100/len(playlist))
        ret_lst.append(elmt)
    return ret_lst
```

#### Build a recommendation off of this data now just to see how it works.

This function will save all unique variables in a playlist to a list

```python
def unique_tracks(playlist):
    ret_lst = list()
    for i in playlist:
        for j in i:
            if j not in ret_lst:
                ret_lst.append(j)
    return ret_lst
```
