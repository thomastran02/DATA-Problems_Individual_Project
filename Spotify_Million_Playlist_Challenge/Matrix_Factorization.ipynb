## Preprocessing

```python
f = open('challenge_set.json','r')
data = json.loads(f.read())
df = pd.json_normalize(data,record_path=['playlists'],meta=['date'])

tracks = df['tracks']

def new_columns(key,col = tracks,empty = ['Empty']):
    info = []
    for i in col:
        if i==[]:
            info.append(empty)
        else:
            lst = list()
            for j in i:
                lst.append(j[key])
            info.append(lst)
    return info

keys = ['track_name','artist_name','track_uri','artist_uri','album_uri','duration_ms','album_name']
for k in keys:
    df[k] = new_columns(k)
```

```python
def unique_tracks(playlist):
    ret_lst = list()
    for i in playlist:
        for j in i:
            if j not in ret_lst:
                ret_lst.append(j)
    return ret_lst
```

```python
def takeout(columns,seed,data):
    np.random.seed(seed)
    ret_lst = list()
    arr = np.random.choice([i for i in range(100)],50,replace=False)
    for c in columns:
        col = list()
        for i in data.index:
            lst = list()
            for a in arr:
                lst.append(data[c][i][a])
            col.append(lst)
        ret_lst.append(col)
    return ret_lst
```

```python
def make_dict(column,values,full_track,data_frame):
    dictionary = dict()
    for track in full_track:
        for index in data_frame.index:
            if track in data_frame[column][index]:
                i = data_frame[column][index].index(track)
                dictionary[track]=data_frame[values][index][i]
                break
    return dictionary
```

```python
def create_ratings(user,tracks,track_col='track_uri',art_col='artist_uri'):
    lst = list()
    for track in tracks:
        if track in user[track_col]:
            lst.append((user[art_col].count(track_artist_uri[track]))/len(user[art_col]))
        else:
            lst.append(0.0)
    return lst
```

```python
def create_matrix(data_frame,tracks,track_col='track_uri',art_col='artist_uri'):
    lst = list()
    for index in range(len(data_frame)):
        lst.append(create_ratings(data_frame.iloc[index],tracks,track_col,art_col))
    arr = np.array(lst)
    return arr
```

## Matrix Factorization

```python
def matrix_factorization(R, P, Q, K, steps = 5000, alpha = 0.001, beta = 0.01):
    
    Q = Q.T
    
    for step in range(steps):
        for i in range(len(R)):
            for j in range(len(R[i])):
                if R[i][j]>0:
                    eij = R[i][j] - np.dot(P[i,:],Q[:,j])
                    
                    for k in range(K):
                        P[i][k] = P[i][k]+alpha*(2*eij*Q[k][j]-beta*P[i][k])
                        Q[k][j] = Q[k][j]+alpha*(2*eij*P[i][k]-beta*Q[k][j])
        eR = np.dot(P,Q)
        e = 0
        for i in range(len(R)):
            for j in range(len(R[i])):
                if R[i][j]>0:
                    e = e + pow(R[i][j]-np.dot(P[i,:],Q[:,j]),2)
        if e<0.01:
            break
    return P, Q.T
```

```python
NoOfUsers = len(mat)
NoOfItems = len(mat[0])
K=10
```

```python
def precision(temp,true):
    count = 0
    num_rec_tracks = 0
    for user in range(len(temp)):
        num_rec_tracks+=len(temp[user])
        for track in temp[user]:
            if track in true[user]:
                count+=1
    return count/num_rec_tracks
```
